```python
import pandas as pd
import numpy as np
import json as js
import geopandas as gpd
import seaborn as sns

from matplotlib import pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# file paths 
district_data_path = "../reference_files/Rice yeild total area.csv"
lulc_data_path = ""
soil_data_path = ""
```

```python
district_df = pd.read_csv(district_data_path)

district_df["Area_Hec_Rice"] = district_df["Area_Hec_Rice"].str.replace(",", "")
district_df["Area_Hec_Rice"] = district_df["Area_Hec_Rice"].astype(float)
```

```python
print(list(district_df["District"].unique()))
```

```python
dist_name = "Lucknow"
temp_df = district_df[district_df["District"]== dist_name]

fig = plt.figure(figsize=(10,6))
sns.lineplot(x=temp_df["year"], y=temp_df["Area_Hec_Rice"], color="g")
ax2 = plt.twinx()
sns.lineplot(x=temp_df["year"], y=temp_df["rice_area_pct"], ax=ax2, color="b")
fig.legend(labels=['Area Hect','', 'Area %', ''])
plt.title("Area for district (and area %): "+dist_name)
plt.show()
```
